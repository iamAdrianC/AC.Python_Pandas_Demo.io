## Number of respondents in each country

```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_java = country_group['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Java').sum())
country_uses_java
```

## Adding strings (respondents and java users) together using concat

```python
java_df = pd.concat([country_respondents, country_uses_java], axis='columns', sort=False)
java_df
```

## Renaming columns

```python
java_df.rename(columns={'Country':'Respondents', 'LanguageHaveWorkedWith':'Java Users'}, inplace=True)
```

```python
java_df
```

## Finding out ratio of java users among the respondents

```python
java_df['PercentKnowsJava'] = (java_df['Java Users']/java_df['Respondents']) * 100
java_df
```

## Sorting out percentages of respondents who know how to use Java

```python
java_df.sort_values(by='PercentKnowsJava', ascending=False, inplace=True)
```
